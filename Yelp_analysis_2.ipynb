```python
import csv
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import operator #need this for sorting dicts later
import matplotlib.cm as cm #for colormapping later
from matplotlib.colors import LogNorm
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

Import custom restauramt and not_restaurant csv files as dataframes

```python
pubs_df = pd.read_csv('pubs.csv')
not_restaurant_df = pd.read_csv('not_restaurant.csv')
pubs_df.head()
pubs_df = pubs_df.drop(['city','Unnamed: 0'], 1)
pubs_df.head()
not_restaurant_df = not_restaurant_df.drop(['Unnamed: 0','city','distance'], 1)
pubs_df.head()
```

```python
not_restaurant_df.head()
```

```python
final_df = pd.DataFrame()
```

```python
df = pd.DataFrame()

pubs1_df = pubs_df.iloc[[0]]

df['LAT_rad'], df['LON_rad'] = np.radians(not_restaurant_df['latitude']), np.radians(not_restaurant_df['longitude'])
df['dLON'] = df['LON_rad'] - math.radians(pubs1_df['longitude'])
df['dLAT'] = df['LAT_rad'] - math.radians(pubs1_df['latitude'])

# using heaversine's formula:
not_restaurant_df['distance'] = 6367 * 2 * np.arcsin(np.sqrt(np.sin(df['dLAT']/2)**2 + math.cos(math.radians(37.2175900)) * np.cos(df['LAT_rad']) * np.sin(df['dLON']/2)**2))
not_restaurant_df = not_restaurant_df.drop_duplicates(subset=['name','postal_code','review_count', 'distance'], keep=False)
measure_1 = not_restaurant_df
measure_1 = measure_1.loc[measure_1['distance'] <=1.0]
mean1 = measure_1['stars'].mean()
density = (measure_1['stars'].count())/ (2*math.pi*1.0*1.0)

final_df = final_df.append({'name': pubs1_df['name'].item(), 'rating': pubs1_df['stars'].item(), 'mean_business_rating': mean1, "density": density}, ignore_index=True)
```
